```python
df1 = pd.read_csv("raw_data.csv")
df1.head()
```

```python
df2 = pd.read_csv("transformed_data.csv")
df2.head()
```

```python
df2.rename({'CODE': 'iso_code','COUNTRY':'location','DATE':'date'}, axis=1)
```

```python
df3=df2.rename({'CODE': 'iso_code','COUNTRY':'location','DATE':'date'}, axis=1)
```

```python
pd.merge(df1,df3,how='left',on='iso_code')
```

```python
df4=pd.merge(df1,df3,how='left',on='iso_code')
```

```python
df4
```

```python
# Top 10 Countries with Highest Covid Cases

df5=df4.groupby('location_x').max().sort_values(by='total_cases',ascending=False)[:10]
df5
```

```python
#Total number of deaths among the countries with the highest number of covid-19 cases

df6=df5['total_deaths'].sum(axis=0)
df6
```

```python
#Percentage of total deaths and total cases among all the countries with the highest number of covid-19 cases
df7=df5['total_deaths'] / df5['total_cases']*100
df7
```

```python
#calculate the death rate of Covid-19 cases
df8=df4['total_deaths'].sum(axis=0)
df8
```

```python
df9=df4['total_cases'].sum(axis=0)
df9
```

```python
df10=df8/df9*100
df10
```

```python
#stringency index-: It is a composite measure of response indicators, including school closures, workplace closures, 
#and travel bans. It shows how strictly countries are following these measures to control the spread of covid-19

df11=df4['stringency_index'].mean()
df11
```
